### **DATA CLEANING**

```python
filter_list = ['QUEENS', 'BRONX','BROOKLYN','STATEN ISLAND','MANHATTAN']
borough_fil =df[df.BOROUGH.isin(filter_list)]
borough_fil
```

```python
df_filter = borough_fil[['ZIP CODE','CRASH DATE','ON STREET NAME','CROSS STREET NAME','BOROUGH','CRASH TIME','NUMBER OF PERSONS INJURED',
       'NUMBER OF PERSONS KILLED', 'NUMBER OF PEDESTRIANS INJURED',
       'NUMBER OF PEDESTRIANS KILLED', 'NUMBER OF CYCLIST INJURED',
       'NUMBER OF CYCLIST KILLED', 'NUMBER OF MOTORIST INJURED',
       'NUMBER OF MOTORIST KILLED','CONTRIBUTING FACTOR VEHICLE 1','VEHICLE TYPE CODE 1', 'VEHICLE TYPE CODE 2',
       'VEHICLE TYPE CODE 3', 'VEHICLE TYPE CODE 4', 'VEHICLE TYPE CODE 5']]
df_filter 
```

```python
queens = df_filter[df_filter.BOROUGH == 'QUEENS']
queens
```

### **RENAMING COLUMNS**

```python
queens.rename(columns = {'CRASH DATE':'CRASH_DATE','ZIP CODE':'ZIP_CODE'}, inplace = True)
queens
```

```python
queens.info()
```

```python
queens.isna().sum()
```

```python
#Was unable to drop null data so simply gave it a value of 1
#When the data is grouped, 1 will be ignored since its a null value 

queens['ZIP_CODE'].fillna(1,inplace=True)
```

```python
queens.groupby(['ZIP_CODE'])['CRASH_DATE'].count().sort_values().iloc[-20:].plot(kind=('bar'))
#Zipcodes were not added together since one is a float 
#dataset was really messy and attempting to drop the decimal places or change the dtype will end up giving a error message
#Below you get multiple 11101 data sets, not giving us the total values we want 
```

```python
#Data could not be merged since some later zipcodes no longer had the decimals, multiple attempts to filter data was 
#made but unsuccessful, so I looked at the top 50 values and from added the zipcodes that didnt add due to the decimal
#Top 10 zipcodes with most accidents in the past 10 years was made 
queens.groupby(['ZIP_CODE'])['CRASH_DATE'].count().sort_values().iloc[-50:]
```
