***Results ***

K-12 schools had the lowest average algebra scores with an average test score of 80.81


```python
import matplotlib.pyplot as plt

school_level= ['k-8','High School','JHS', 'k-12','Secondary School']
average_alg_score = [83.758,83.29,84.12,80.81,82.04]

plt.scatter(school_level, average_alg_score)
plt.title('School Level Average Common Algebra Scores')
plt.xlabel('School Level')
plt.ylabel('Average Alg Score')
plt.grid(True)
plt.show()
```
