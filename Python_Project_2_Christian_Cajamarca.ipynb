**Python Project 2**

For this project we wanted compare the average wait times depending on race. 

**Purpose of project** 
  - See if there is any trend 
  - Is there any bias going on in the hospital 
  - Which age groups tend to wait more 
  - Compare Hispanic, African American, and Caucasian wait times 

**Steps to filter and get the desired dataframes**
  - Drop the columns that were unnecessary 
  - Filter the necessary columns and create a new dataframe 
  - Examine unique data within each columns
  - Rename Columns
  - Create new dataframes for Hispanic, African American and Caucasian 
  - Plot each dataset by grouping based on age and taking the mean wait time 
  for each age group



```python
ave_hispanic = hispanic.groupby('Age')['Time_In_Hospital'].mean()
```

```python
ave_hispanic.plot(kind='bar',
          figsize=(10,5),           
          title='Hispanics Average Wait Times x Age',
          ylabel= 'Average_Wait_Time') 
```

```python
ave_caucasian = caucasian.groupby('Age')['Time_In_Hospital'].mean()
```

```python
ave_caucasian.plot(kind='bar',
          figsize=(10,5),           
          title='Caucasian Average Wait Times x Age',
          ylabel= 'Average_Wait_Time') 
```

```python
ave_africanamerican = africanamerican.groupby('Age')['Time_In_Hospital'].mean()
```

```python
ave_africanamerican.plot(kind='bar',
          figsize=(10,5),           
          title='African American Average Wait Times x Age',
          ylabel= 'Average_Wait_Time') 
```

```python
ave_africanamerican.plot(kind='line',  
          linewidth=2,
          label = 'African_American')
ave_caucasian.plot(kind='line',
          linewidth=3,
          label = 'Caucasian') 
ave_hispanic.plot(kind='line',
          figsize=(14,8), 
          linewidth=3,          
          color='red',
          label = 'Hispanic',
          title='Comparing_Average_Wait_Time',
          ylabel= 'Average_Wait_Time')

plt.legend()
```

Results 

- Based on the average wait times based on age group: 
  - African Americans are more likely to wait longer 
  - Hispanics are least likely to have a longer wait time 
  - For all races the older you get the longer you are going to be waiting in the hospital 
  - At age 80+ hispanic wait time drastically increases 

- Any Bias
  - From the age of 15 y.o to 75 y.o hispanics and caucasians have a similar wait time 
  - From the age of 15 y.o to 75 y.o african americans have a longer wait time about a 30-40 min longer wait 
  - Its hard to make a claim of bias since we have no health indicators that explain a persons health condition. Due to hospital privacy we were unable to access that dataframe.
